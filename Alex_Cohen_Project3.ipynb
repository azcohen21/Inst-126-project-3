```python
# open the userlog file
##LINK TO GITHUB REPO: https://github.com/azcohen21/Inst-126-project-3
f = open("userlog.log")

# reading the userlog text and storing in variable "text"
text = f.readlines()
#print(text)
dates_by_user = {}

#for loop
for line in text:
    #set variable for each user as a list of events.
    
    #split the data, store each line as a list called "event"
    event = line.split()
    
    #to get user name
    user = event[4]
    #if in dict, get value (dict) for user, if not, create dict (value)
    
    if user not in dates_by_user:
        dates_by_user[user] = {}
        
    
    user_event_date = event[0]
    if user_event_date in dates_by_user[user]:
        dates_by_user[user][user_event_date].append(event)
    else: 
        dates_by_user[user][user_event_date] = [event]
        
    
    
#new dictionary "sus"
filename = "suspicious_report.txt"
sus = {}
#users in dates_by_user where key is user and the value is that date
# for each user
for user, dates in dates_by_user.items():
    # ('username', {'date': [event-list]})
    #for each day
    for day, events_list in dates.items():
        # count logins
        logins = 0
        logouts = 0
        early_login = False
        for event in events_list:
            login_hour = event[1].split(":")[0]
            
            if event[2] == 'login':
                logins += 1
                
                if int(login_hour) < 5:
                    early_login = True
                    
            
        if logins > 5 or early_login:
            if user not in sus:
                #adds user to sus dict
                sus[user] = {}
            if day not in sus[user]:
                sus[user][day] = events_list
                
#print(total_count)

total_check = 0
for dates in sus.values():
    total_check += len(dates) 
with open(filename, 'w') as file_object:

    file_object.write(f"   ============================== \n=== Suspicious Behavior ({total_check} Cases) === \n   ==============================\n")    
    for user, dates in sus.items():
        file_object.write(f"\n\n")
        file_object.write(f"{user}\t{len(dates)}\n")
        for day in sorted(dates):
            events_list = dates[day]
            file_object.write(f"\tDATE: [{day}] ---\n")
            for event in sorted(events_list):
                if event[2] == "login":
                    file_object.write(f"\t\t{event[1]}\t{event[2]}\t\t{event[3]}\n")
                if event[2] == "logout":
                    file_object.write(f"\t\t{event[1]}\t{event[2]}\t\t\t{event[3]}\n")

                #print(f"\n")
        # if logins > 5 count as sus
        #Export to file

#new dictionary "irresponsible"
filename = "irresponsible_report.txt"

irresponsible = {}

#users in dates_by_user key is user and the value is that date
# for each user
for user, dates in dates_by_user.items():
    # ('username', {'date': [event-list]})
    #for each day
    for day, events_list in dates.items():
        # count logins
        logins = 0
        logouts = 0
        for event in events_list:
            if event[2] == 'login':
                logins += 1
            elif event[2] == 'logout':
                logouts +=1
        if logins > logouts:
            if user not in irresponsible:
                #adds user to irresponsible dict
                irresponsible[user] = {}
            if day not in irresponsible[user]:
                irresponsible[user][day] = events_list
            
                
total_check = 0
for dates in irresponsible.values():
    total_check += len(dates)
with open(filename, 'w') as file_object:

    file_object.write(f"   ============================== \n=== Irresponsible Behavior ({total_check} Cases) === \n   ==============================\n")    

    for user, dates in irresponsible.items(): 
        file_object.write(f"\n\n")
        file_object.write(f"{user}\t{len(dates)}\n")
        for day in sorted(dates):
            events_list = dates[day]
            file_object.write(f"\t\tDATE: [{day}] ---\n")
            for event in sorted(events_list):
                if event[2] == "login":
                    file_object.write(f"\t\t\t{event[1]}\t{event[2]}\t\t{event[3]}\n")
                if event[2] == "logout":
                    file_object.write(f"\t\t\t{event[1]}\t{event[2]}\t\t\t{event[3]}")
                    #print(f"\n")
            # if logins < logouts then count as irresponsible
            #Export to file

#new dictionary "glitches"
filename = "glitch_report.txt"

glitches = {}

#users in dates_by_user key is user and the value is that date
# for each user
for user, dates in dates_by_user.items():
    # ('username', {'date': [event-list]})
    #for each day
    for day, events_list in dates.items():
        # count logins
        logins = 0
        logouts = 0
        for event in events_list:
            if event[2] == 'login':
                logins += 1
            elif event[2] == 'logout':
                logouts +=1
        if logins < logouts:
            if user not in glitches:
                #adds user to glitches dict
                glitches[user] = {}
            if day not in glitches[user]:
                glitches[user][day] = events_list
            
total_check = 0
for dates in glitches.values():
    total_check += len(dates)
with open(filename, 'w') as file_object:
    file_object.write(f"   ============================== \n=== Glitches ({total_check} Cases) === \n   ==============================\n")                    
    for user, dates in glitches.items():
        file_object.write(f"\n\n")
        file_object.write(f"{user}\t{len(dates)}\n")
        for day in sorted(dates):
            events_list = dates[day]
            file_object.write(f"\t\tDATE: [{day}] ---\n")
            for event in sorted(events_list):
                if event[2] == "login":
                    file_object.write(f"\t\t\t{event[1]}\t{event[2]}\t\t{event[3]}\n")
                if event[2] == "logout":
                    file_object.write(f"\t\t\t{event[1]}\t{event[2]}\t\t\t{event[3]}\n")
                    #print(f"\n")
                    
#new dictionary "domains"
filename = "domain_report.txt"
domains = {}

#users in dates_by_user key is user and the value is that date
for user in dates_by_user:
    user_parts = user.split("@")
    user_name = user_parts[0]
    domain = user_parts[1]
    
         
        #need number of users per domain
    if domain not in domains:
        #setting key for domains dict. key is domain
        domains[domain] = set()
    domains[domain].add(user_name)                

with open(filename, 'w') as file_object:
    file_object.write(f"===================================\n===DOMAINS AND NUMBER OF USERS PER DOMAIN===\n=================================== \nTotal # of domains: {len(domains)} \n===================================")
    file_object.write("\n")
    for domain, users_set in domains.items():
        file_object.write(f"{domain} {len(users_set)}\n")
        
```
